> **Chapter 6**. 이벤트 처리하기 : React Events

리엑트의 이벤트 :  카멜케이스 사용, JSX사용해서 함수로 이벤트 핸들러 전달

돔 이벤트 : 소문자 사용, 문자열로 이벤트 핸들러 전달

리엑트에서 지원하는 이벤트 종류
- onClick
- onChange
- onInput
- onFocus
- onMouseEnter
- onMouseLeave



```python

function App() {
  const handleOnMouseEnter = () => {
    console.log("안녕하세요!");
  };

  const handleOnMouseLeave = () => {
    console.log("안녕히가세요!");
  };

  return (
    <div>
      <p 
        onMouseEnter={handleOnMouseEnter} 
        onMouseLeave={handleOnMouseLeave}
      >
        여기에 마우스를 올려보세요!
      </p>
    </div>
  );
}

export default App;

" -> 이벤트 사용예시 "
```
